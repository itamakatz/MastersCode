```python
def get_l_star(pi_0,s_0,s_1):
    if(pi_0 < 0.5): # depending on pi_0 if s_a,s_b are s_0,s_1 or s_1,s_0
        return ((s_0*pi_0)-(s_1*(1-pi_0)))/((1-pi_0)-pi_0)
    else:
        return ((s_1*pi_0)-(s_0*(1-pi_0)))/((1-pi_0)-pi_0)

def get_delta_0(pi_0,s_0,s_1):
    return pi_0*(1-pi_0)/(pi_0-(1-pi_0))*(s_0-s_1)**2;

def get_lambda_i(s_i,l):
    return (s_i + l)**2

def get_lambda_bar(pi_0,lambda_0,lambda_1):
    return lambda_0*pi_0 + lambda_1*(1-pi_0)
            
def get_pi_n(pi_o,lambda_0,lambda_bar):
    return pi_0*lambda_0/lambda_bar

def get_lambda_bar_yuval_1(pi_0,pi_n,delta_0):
    return delta_0 * (math.sqrt(pi_n*(1-pi_0)) + math.sqrt((1-pi_n)*pi_0))**2
def get_lambda_bar_yuval_2(pi_0,pi_n,delta_0):
    return delta_0 * (math.sqrt(pi_n*(1-pi_0)) - math.sqrt((1-pi_n)*pi_0))**2

def get_lambda_bar_itamar_1(pi_0,pi_n,delta_0):
    lambda_bar_yuval_1 = get_lambda_bar_yuval_1(pi_0,pi_n,delta_0)
    return lambda_bar_yuval_1*(pi_0-(1-pi_0)) / ((pi_n-pi_0)**2)
def get_lambda_bar_itamar_2(pi_0,pi_n,delta_0):
    lambda_bar_yuval_2 = get_lambda_bar_yuval_2(pi_0,pi_n,delta_0)
    return lambda_bar_yuval_2*(pi_0-(1-pi_0)) / ((pi_n-pi_0)**2)

def get_g_pi_pi(pi_0,pi_n):
    pi_0*(1-pi_0)*pi_n
    lambda_bar_yuval_2 = get_lambda_bar_yuval_2(pi_0,pi_n,delta_0)
    return lambda_bar_yuval_2*(pi_0-(1-pi_0)) / ((pi_n-pi_0)**2)
```
